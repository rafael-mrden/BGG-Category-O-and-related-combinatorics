```python
f=open("Graded_characters_in_O_0_data/miscellaneous/D4_right_order.txt", "r")  # Mind the type here!
R_order_string =f.read()
f.close()

def R_smaller(x,y):
    if "(%s, %s)"%(x,y) in R_order_string:
        return True
    else:
        return False
    

D = # Duflo_Involutions() or copy from file
    
elem = # copy from file "?? elementary isomorphisms and decompositions for computer.txt".

LC_123 = # copy from file "left cells.txt".

Left_Cells = [ [convert_from_123(x) for x in comp] for comp in LC_123 ]

def L_cell(d):
    for cel in Left_Cells:
        if d in cel:
            return cel
        
def R_cell(d):
    for cel in Left_Cells:
        if d.inverse() in cel:
            return [x.inverse() for x in cel]

def two_cell(w):
    Lw = L_cell(w)
    result = []
    
    for x in Lw:
        Rx = R_cell(x)
        for y in Rx:
            if y not in result:
                result.append(y)
    return result
```

```python
def print_equation(equation):
    string1 = ""
    for p in equation[0]:
        string1 += " + M(%s,%s)"%( convert_to_123(p[0]) , convert_to_123(p[1]) )        

    string2 = ""
    for p in equation[1]:
        string2 += " + M(%s,%s)"%( convert_to_123(p[0]) , convert_to_123(p[1]) )
        
    string3 = string1[3:] + " == " + string2[3:]
    print(string3)
```

```python
elem22_cancelation = []

for eq in elem22:
    
    cont = 0
    
    [A,B] = eq[0]
    [C,D_] = eq[1]
    AA = conn_comp(A)
    BB = conn_comp(B)
    CC = conn_comp(C)
    DD = conn_comp(D_)
    
    if AA == CC and BB != DD:
        eq2 = [{B},{D_}]
        cont = 1
    if AA == DD and BB != CC:
        eq2 = [{B},{C}]
        cont = 1
    if BB == CC and AA != DD:
        eq2 = [{A},{D_}]
        cont = 1
    if BB == DD and AA != CC:
        eq2 = [{A},{C}]
        cont = 1
    
    if cont == 1:
    
        if (eq2 not in elem22_cancelation) and (list(reversed(eq2)) not in elem22_cancelation):
            elem22_cancelation.append(eq2)
            print_equation(eq2)
```
