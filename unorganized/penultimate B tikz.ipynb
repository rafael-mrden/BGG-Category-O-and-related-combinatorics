```python
k = 3
margin = 0.175
margin2 = 0.05


print("\\begin{tikzpicture}[scale=1, yscale=-1]")

for i in range(k+3):
    for j in range(k+3):
        if i != k+1 and j != k+1:
            print("\\fill [gray!20] (%.2f,%.2f) rectangle (%.2f,%.2f);"
                  %(2*i-0.5+margin2, 2*j-0.5+margin2, 2*i+1+0.5-margin2, 2*j+1+0.5-margin2))

print("\\newcommand{\\arrow}[4]{\\draw [->] (%.2f*#1+%.2f*#3,%.2f*#2+%.2f*#4) -- (%.2f*#1+%.2f*#3,%.2f*#2+%.2f*#4);}"%(1-margin,margin, 1-margin, margin, margin, 1-margin, margin, 1-margin))

print("\\node[] at (0,0) {$u_{00}$};")
print("\\node[] at (1,1) {$w_{00}$};")

for i in range(1,k+1):
    print("\\node[] at (%.2f,0.5) {$u_{0%d}$};"%(2*i+0.5,i))

for j in range(1,k+1):
    print("\\node[] at (0.5,%.2f) {$u_{%d0}$};"%(2*j+0.5,j))
    
for i in range(1,k+1):
    for j in range(1,k+1):
        print("\\node[] at (%.2f,%.2f) {$u_{%d%d}$};"%(2*i,2*j,j,i))
        print("\\node[] at (%.2f,%.2f) {$w_{%d%d}$};"%(2*i+1,2*j+1,j,i))
        
print("\\node[] at (%.2f,0.5) {$\\cdots$};"%(2*(k+1)+0.5))
print("\\node[] at (0.5,%.2f) {$\\vdots$};"%(2*(k+1)+0.5))  

for i in range(1,k+1):
    print("\\node[] at (%.2f,%.2f) {$\\cdots$};"%(2*(k+1), 2*i))
    print("\\node[] at (%.2f,%.2f) {$\\cdots$};"%(2*(k+1)+1, 2*i+1))

for j in range(1,k+1):
    print("\\node[] at (%.2f,%.2f) {$\\vdots$};"%(2*j, 2*(k+1)))
    print("\\node[] at (%.2f,%.2f) {$\\vdots$};"%(2*j+1, 2*(k+1)+1))

print("\\node[] at (%.2f,%.2f) {$\\ddots$};"%(2*(k+1), 2*(k+1)))
print("\\node[] at (%.2f,%.2f) {$\\ddots$};"%(2*(k+1)+1, 2*(k+1)+1))

print("\\node[] at (%.2f,0.5) {$u_{0n}$};"%(2*(k+2)+0.5))
print("\\node[] at (0.5,%.2f) {$u_{n0}$};"%(2*(k+2)+0.5))

for i in range(1,k+1):
    print("\\node[] at (%.2f,%.2f) {$u_{%dn}$};"%(2*(k+2), 2*i,i))
    print("\\node[] at (%.2f,%.2f) {$w_{%dn}$};"%(2*(k+2)+1, 2*i+1,i))

for j in range(1,k+1):
    print("\\node[] at (%.2f,%.2f) {$u_{n%d}$};"%(2*j, 2*(k+2), j))
    print("\\node[] at (%.2f,%.2f) {$w_{n%d}$};"%(2*j+1, 2*(k+2)+1, j))

print("\\node[] at (%.2f,%.2f) {$\\cdots$};"%(2*(k+1), 2*(k+2)))
print("\\node[] at (%.2f,%.2f) {$\\cdots$};"%(2*(k+1)+1, 2*(k+2)+1))    

print("\\node[] at (%.2f,%.2f) {$\\vdots$};"%(2*(k+2), 2*(k+1)))
print("\\node[] at (%.2f,%.2f) {$\\vdots$};"%(2*(k+2)+1, 2*(k+1)+1))

print("\\node[] at (%.2f,%.2f) {$u_{nn}$};"%(2*(k+2), 2*(k+2)))
print("\\node[] at (%.2f,%.2f) {$w_{nn}$};"%(2*(k+2)+1, 2*(k+2)+1)) 

print("\\arrow{0}{0}{0.5}{2.5};")
print("\\arrow{0}{0}{2.5}{0.5};")
print("\\arrow{0.5}{2.5}{1}{1};")
print("\\arrow{2.5}{0.5}{1}{1};")

for i in range(1,k+2):
    print("\\arrow{%.2f}{0.5}{%.2f}{0.5};"%(2*(i+1)+0.5,2*i+0.5))

for j in range(1,k+2):
    print("\\arrow{0.5}{%.2f}{0.5}{%.2f};"%(2*(j+1)+0.5,2*j+0.5))
    
for i in range(1,k+1):
    for j in range(i,k+2):
        print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(j+1) +1, 2*i+1 , 2*j +1, 2*i+1 ))
    for j in range(1,i):
        print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*j +1, 2*i+1 , 2*(j+1) +1, 2*i+1 ))    
    for j in range(1,k+2):
        print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(j+1) , 2*i , 2*j , 2*i ))
        
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2 , 2*i , 0.5 , 2*i+0.5 ))  
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 0.5 , 2*i+0.5, 3 , 2*i +1 ))

for i in range(1,k+1):
    for j in range(i,k+2):
        print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*i+1, 2*(j+1) +1 , 2*i+1 , 2*j +1 ))
    for j in range(1,i):
        print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%(2*i+1 ,  2*j +1, 2*i+1, 2*(j+1) +1 ))      
    for j in range(1,k+2):
        print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*i , 2*(j+1) , 2*i, 2*j ))    
        
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*i , 2 ,  2*i+0.5 , 0.5 ))  
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*i+0.5, 0.5 , 2*i +1 , 3 ))  
    
for j in range(1,k+2):
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(k+2) , 2*(j+1) ,  2*(k+2) , 2*(j) ))    
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(k+2)+1 , 2*(j)+1 ,  2*(k+2)+1 , 2*(j+1)+1 ))  
     
print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2 , 2*(k+2) , 0.5 , 2*(k+2)+0.5 ))  
print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 0.5 , 2*(k+2)+0.5, 3 , 2*(k+2) +1 ))

for j in range(1,k+2):
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(j+1) ,  2*(k+2) , 2*(j) , 2*(k+2) ))    
    print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(j)+1 , 2*(k+2)+1 , 2*(j+1)+1 , 2*(k+2)+1 ))  

print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(k+2) , 2 ,  2*(k+2)+0.5 , 0.5 ))  
print("\\arrow{%.2f}{%.2f}{%.2f}{%.2f};"%( 2*(k+2)+0.5, 0.5 , 2*(k+2) +1 , 3 ))  


print("\\end{tikzpicture}")
```
